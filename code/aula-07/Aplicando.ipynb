## Funes base

```python
from pathlib import Path


def print_file(file_path: Path):
    if file_path.is_file():
        # Open and read the file
        with file_path.open(mode='r') as file:
            file_content = file.read()
        
        print(file_content)
    else:
        print("O arquivo no existe.")

def write_to_file(file_path: Path, file_content: str):
    with file_path.open(mode='w') as file:
        file.write(file_content)
```

## Constantes

```python
file_path = Path("arquivo.txt")

greeting = """
Boa noite!

Este  meu arquivo.txt criado em Python

Abrao!
"""
```

## Testes

```python
print("Antes da escrita:\n")
print_file(file_path)
```

```python
write_to_file(file_path, greeting)

print("O arquivo aps a criao :\n")
print_file(file_path)
```
